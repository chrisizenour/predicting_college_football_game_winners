# Appendix A. `model_table` Feature Sets and Data Dictionary

## Purpose
This appendix documents the feature families used in the season-rollover winner-prediction analysis and provides a column-by-column data dictionary for the final `model_table` dataframe. The table is framed from the home-team perspective: for most engineered differential variables, positive values indicate an advantage for the home team.

## Feature set overview

- **Baseline / context**: Captures simple game context and ranking status. This set asks how far prediction can go using neutral-site context, conference status, ranking indicators, and how much current-season history each team has entering the game.
- **Talent / roster priors**: Captures pregame roster quality using 247-derived recruiting, overall, and transfer ratings. This set asks how much signal exists from talent and roster construction before relying on current-season box-score performance.
- **Performance / team-quality**: Captures season-to-date and recent-form football performance using prior games only. These variables summarize how well each team has played entering the matchup.
- **Matchup**: Captures offense-versus-defense fit. These variables ask whether one team’s offensive strengths line up favorably against the other team’s defensive weaknesses.

## Feature family membership

| Feature family | Columns in analysis tier |
|---|---|
| Baseline / context | `neutral`, `conference_game`, `ranked_home`, `ranked_away`, `both_ranked`, `home_pre_games_played_pre`, `away_pre_games_played_pre`, `rank_advantage_home` |
| Talent / roster priors | `recruiting_avg_4yr_diff`, `recruiting_points_mean_4yr_diff`, `overall_avg_diff`, `overall_points_diff`, `transfer_avg_diff`, `transfer_points_diff`, `home_pre_overall_avg_missing`, `home_pre_transfer_avg_missing`, `away_pre_overall_avg_missing`, `away_pre_transfer_avg_missing` |
| Performance / team-quality | `win_pct_diff`, `margin_pre_mean_diff`, `margin_pre_last3_diff`, `points_for_pre_mean_diff`, `points_against_pre_mean_diff`, `ypp_for_pre_mean_diff`, `ypp_against_pre_mean_diff`, `yppa_for_pre_mean_diff`, `ypra_for_pre_mean_diff`, `turnover_margin_pre_mean_diff`, `third_down_rate_for_pre_mean_diff`, `third_down_rate_against_pre_mean_diff`, `possession_for_pre_mean_diff` |
| Matchup | `pass_matchup_diff`, `rush_matchup_diff`, `ypp_matchup_diff`, `opp_pass_matchup_diff`, `opp_rush_matchup_diff`, `opp_ypp_matchup_diff` |

## Data dictionary for `model_table`

Column roles:
- **Identifier** = descriptive field used to locate a game
- **Target** = outcome field
- **Predictor** = pregame model input

Interpretation notes are written from the home-team perspective.

| Column | Role | Definition | Interpretation / modeling note |
|---|---|---|---|
| `season` | Identifier | Football season for the game. | Used for season-rollover train/test splits and time ordering. |
| `week` | Identifier | Week number of the game. | Useful for ordering and phase-of-season analysis. |
| `date` | Identifier | Calendar date of the game. | Used for chronological ordering. |
| `away` | Identifier | Normalized away-team name. | Describes the away team in the matchup. |
| `home` | Identifier | Normalized home-team name. | Describes the home team in the matchup. |
| `neutral` | Predictor | Neutral-site indicator (`1 = neutral`, `0 = not neutral`). | A positive model effect would mean neutral-site status increases home-win odds; often it weakens home-field advantage. |
| `conference_game` | Predictor | Same-conference indicator (`1 = same conference`, `0 = different conferences`). | Lets the model learn whether conference familiarity changes win probabilities. |
| `ranked_home` | Predictor | Indicator that the home team was ranked entering the game. | Captures ranked-versus-unranked context. |
| `ranked_away` | Predictor | Indicator that the away team was ranked entering the game. | Captures away-team ranking context. |
| `both_ranked` | Predictor | Indicator that both teams were ranked entering the game. | Flags games between two ranked teams. |
| `margin_home` | Target | Final scoring margin from the home-team perspective. | Positive values mean the home team won by that margin; negative values mean the away team won. |
| `home_win` | Target | Binary game winner target (`1 = home win`, `0 = away win`). | Primary classification target for winner-prediction modeling. |
| `home_pre_games_played_pre` | Predictor | Number of prior games the home team had already played that season. | Measures how much current-season information is available for the home team. |
| `away_pre_games_played_pre` | Predictor | Number of prior games the away team had already played that season. | Measures how much current-season information is available for the away team. |
| `rank_advantage_home` | Predictor | Ranking advantage for the home team, defined as `rank_away_filled - rank_home_filled`. | Positive values mean the home team is better ranked. |
| `win_pct_diff` | Predictor | Difference in prior season-to-date win percentage: home minus away. | Positive values favor the home team. |
| `margin_pre_mean_diff` | Predictor | Difference in average prior scoring margin across all previous games: home minus away. | Positive values favor the home team. |
| `margin_pre_last3_diff` | Predictor | Difference in average scoring margin over each team’s previous three games: home minus away. | Positive values indicate better recent form for the home team. |
| `recruiting_avg_4yr_diff` | Predictor | Difference in four-year mean recruiting class average rating: home minus away. | Positive values indicate stronger average recruiting quality for the home team. |
| `recruiting_points_mean_4yr_diff` | Predictor | Difference in four-year mean recruiting points: home minus away. | Positive values indicate stronger recruiting-point profile for the home team. |
| `overall_avg_diff` | Predictor | Difference in overall team rating average from 247-derived season data: home minus away. | Positive values indicate a stronger overall rating for the home team. |
| `overall_points_diff` | Predictor | Difference in overall team rating points: home minus away. | Positive values indicate a stronger overall-points rating for the home team. |
| `transfer_avg_diff` | Predictor | Difference in transfer rating average: home minus away. | Positive values indicate stronger transfer quality for the home team. |
| `transfer_points_diff` | Predictor | Difference in transfer rating points: home minus away. | Positive values indicate stronger transfer-point profile for the home team. |
| `points_for_pre_mean_diff` | Predictor | Difference in average points scored entering the game: home minus away. | Positive values mean the home team had been scoring more. |
| `points_against_pre_mean_diff` | Predictor | Difference in average points allowed entering the game: home minus away. | Positive values usually hurt the home team because they indicate the home team had allowed more points. |
| `ypp_for_pre_mean_diff` | Predictor | Difference in offensive yards per play entering the game: home minus away. | Positive values favor the home team. |
| `ypp_against_pre_mean_diff` | Predictor | Difference in defensive yards per play allowed entering the game: home minus away. | Positive values usually hurt the home team because they indicate the home defense had allowed more efficiency. |
| `yppa_for_pre_mean_diff` | Predictor | Difference in offensive yards per pass attempt entering the game: home minus away. | Positive values favor the home team. |
| `ypra_for_pre_mean_diff` | Predictor | Difference in offensive yards per rush attempt entering the game: home minus away. | Positive values favor the home team. |
| `turnover_margin_pre_mean_diff` | Predictor | Difference in average turnover margin entering the game: home minus away. | Positive values favor the home team. |
| `third_down_rate_for_pre_mean_diff` | Predictor | Difference in offensive third-down conversion rate entering the game: home minus away. | Positive values favor the home team. |
| `third_down_rate_against_pre_mean_diff` | Predictor | Difference in defensive third-down conversion rate allowed entering the game: home minus away. | Positive values usually hurt the home team because they indicate the home defense allowed a higher conversion rate. |
| `possession_for_pre_mean_diff` | Predictor | Difference in average time of possession entering the game: home minus away. | Positive values mean the home team had controlled possession longer on average. |
| `pass_matchup_diff` | Predictor | Home passing offense pregame average minus away passing defense allowed pregame average. | Positive values suggest a favorable home passing matchup. |
| `rush_matchup_diff` | Predictor | Home rushing offense pregame average minus away rushing defense allowed pregame average. | Positive values suggest a favorable home rushing matchup. |
| `ypp_matchup_diff` | Predictor | Home yards-per-play offense pregame average minus away yards-per-play defense allowed pregame average. | Positive values suggest a favorable overall efficiency matchup for the home team. |
| `opp_pass_matchup_diff` | Predictor | Away passing offense pregame average minus home passing defense allowed pregame average. | Positive values usually hurt the home team because they suggest a favorable away passing matchup. |
| `opp_rush_matchup_diff` | Predictor | Away rushing offense pregame average minus home rushing defense allowed pregame average. | Positive values usually hurt the home team because they suggest a favorable away rushing matchup. |
| `opp_ypp_matchup_diff` | Predictor | Away yards-per-play offense pregame average minus home yards-per-play defense allowed pregame average. | Positive values usually hurt the home team because they suggest a favorable away overall-efficiency matchup. |
| `home_pre_overall_avg_missing` | Predictor | Indicator that the home team’s `overall_avg` value was missing before imputation. | Helps the model distinguish missingness from a true low value. |
| `home_pre_transfer_avg_missing` | Predictor | Indicator that the home team’s `transfer_avg` value was missing before imputation. | Helps the model distinguish missingness from a true low value. |
| `away_pre_overall_avg_missing` | Predictor | Indicator that the away team’s `overall_avg` value was missing before imputation. | Helps the model distinguish missingness from a true low value. |
| `away_pre_transfer_avg_missing` | Predictor | Indicator that the away team’s `transfer_avg` value was missing before imputation. | Helps the model distinguish missingness from a true low value. |

## Interpretation tips

- Most differential variables are defined as `home minus away`. Positive values usually favor the home team.
- Some defensive-prevention and opponent-matchup variables work in the opposite direction. For example, larger positive values for `points_against_pre_mean_diff`, `ypp_against_pre_mean_diff`, or the `opp_*` matchup variables are usually unfavorable for the home team.
- The outcome columns `margin_home` and `home_win` are targets and should not be used as predictors when training winner-prediction models.
